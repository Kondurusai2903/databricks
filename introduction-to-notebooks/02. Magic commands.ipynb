# Databricks Magic Commands Revision 

## Language Magic Commands
- `%python` — run Python code (default in Python notebooks)
- `%scala` — run Scala code
- `%r` — run R code
- `%sql` — run SQL queries directly on tables/views

## Documentation
- `%md` — render Markdown for documentation, headers, notes within the notebook

## File System
- `%fs` — shell-style wrapper for file system operations (list, copy, move, etc.)
  - e.g. `%fs ls /mnt/data`
  - Under the hood, this maps to `dbutils.fs` commands
- `dbutils.fs` — the programmatic module version (more flexible, usable inside Python/Scala code, not just standalone commands)

## Shell
- `%sh` — run shell/bash commands
  - ⚠️ Runs ONLY on the driver node, not distributed across the cluster

## Package Management
- `%pip` — install Python packages (scoped to the notebook session)
  - e.g. `%pip install pandas`
  - Replaced the older `dbutils.library` approach

## Notebook Chaining
- `%run` — run another notebook and IMPORT its variables/functions into current scope (shared state)
  - e.g. `%run ./utils_notebook`
  - ⚠️ Different from `dbutils.notebook.run()` which runs in an isolated context and does NOT share variables

## Exam Gotchas
- `%fs` vs `dbutils.fs` — same underlying operations, `%fs` is just the shorthand magic command version
- `%run` vs `dbutils.notebook.run()` — `%run` shares variables/state (like an import), `dbutils.notebook.run()` runs isolated and only returns an exit value
- `%sh` is driver-only — can't use it to run something across all worker nodes

# Magic commands 

- **%python**,**%scala**,**%sql**,**%r** : Switch to a different language for a specific cell
- **%md** : markdown for documenting notebooks
- **%fs** : run file system commands
- **%sh** : Run shell commands(Driver node only)
- **%pip** : install Python libraries
- **%run** : include/import another notebook into current notebook


### %fs : Run File System `Commands`


```python
# Examples for all dbutils.fs commands

# List directory contents
dbutils.fs.ls('/')

# Create a directory
dbutils.fs.mkdirs('file:/tmp/example_dir')

# Copy a file or directory
dbutils.fs.cp('file:/tmp/example_dir', 'file:/tmp/example_dir_copy', recurse=True)

# Move a file or directory
dbutils.fs.mv('file:/tmp/example_dir_copy', 'file:/tmp/example_dir_moved', recurse=True)

# Remove a file or directory
dbutils.fs.rm('file:/tmp/example_dir_moved', recurse=True)

# Write a string to a file
dbutils.fs.put('file:/tmp/example_file.txt', 'Hello Databricks!', overwrite=True)

# Read the first bytes of a file
dbutils.fs.head('file:/tmp/example_file.txt', max_bytes=10)

# Mount a directory (example, requires source and configs)
# dbutils.fs.mount(source="dbfs:/mnt/source", mount_point="/mnt/mount_point", extra_configs={})

# List all mounts
dbutils.fs.mounts()

# Unmount a mount point
# dbutils.fs.unmount("/mnt/mount_point")

# Update an existing mount (example, requires source and configs)
# dbutils.fs.updateMount(source="dbfs:/mnt/source", mount_point="/mnt/mount_point", extra_configs={})

# Refresh mount cache
dbutils.fs.refreshMounts()

# Show help for dbutils.fs
dbutils.fs.help()
```
