
### Higher Order Functions

- Higher order Functions are the functions that operate on complex data types such as arrays and maps
- They allow you to pass functions as argument (such as lambda expressions), apply transformations and return arrays or maps
- They are extremely useful for manipulating arrays without exploding them 

#### Commonly used Higher Order Array Functions

- TRANSFORM
- FILTER
- EXISTS
- AGGREGATE


##### Syntax

`<function_name> (array_column, lambda expression)`

_lambda expression_`element -> expression`

```python
%sql
CREATE OR REPLACE TEMPORARY VIEW order_items
as 
SELECT * FROM 
VALUES 
   (1, array('smartphone','laptop','monitor')),
   (2, array('tablet','headphones','smartwatch')),
   (3, array('keyboard','mouse'))
AS orders(order_id,items);
```

```python
%sql
select * from order_items;
```

Convert all the item names to UPPERCASE(Transform Function)

```python
%sql
select order_id,transform(items, x -> upper(x)) as upper_items from order_items;
```

Filter Only items that contains the string 'smart' 

```python
%sql
select order_id,filter(items, x -> x LIKE '%smart%') as smart_str from order_items;
```

check to see whether the orders includes any 'monitor'

```python
%sql
select order_id, exists(items, x -> x = 'monitor') as exists from order_items;
```

## Map Functions


A map is like a key-value pair like a dictionary

`{ 'laptop' : 1200, 'phone': 699}`

### Commonly used Higher Order Map Functions

- TRANSFORM_VALUES
- TRANSFORM_KEYS
- MAP_FILTER

##### syntax

`<function_name> (map_column, lambda expression) `

_lambda expression:_ `(key,value) -> expression`

```python
%sql
CREATE OR REPLACE TEMPORARY VIEW order_item_prices
AS
SELECT * FROM 
VALUES 
    (1, map('smartphone',699, 'laptop',999, 'monitor',199)),
    (2, map('tablet',299, 'headphones',99, 'smartwatch',199)),
    (3, map('keyboard',49, 'mouse',29))
as orders(order_id,items);
```

```python
%sql
select * from order_item_prices;
```

convert all the item names to be UPPERCASE(TRANSFORM_KEYS Function)

```python
%sql
select order_id, transform_keys(items, (key,value) -> upper(key)) as items_upper_case from order_item_prices;
```

```python
%sql
select order_id, transform_values(items, (key,value) -> round(value * 1.10,2)) as items_upper_case from order_item_prices;
```

```python
%sql
select order_id, map_filter(items, (item,price) -> price > 250 ) as premium_items from order_item_prices;
```
