
### Extract Data - using Pyspark

- Run Sql commands using python - spark.sql.Function
- Spark DataFrame Reader API
- Reading Tables using spark.table.function


#### Run Sql commands using python - spark.sql.Function

```python
df = spark.sql('select * from json.`/Volumes/gizmobox/landing/operational_data/customers`')
display(df)
```

```python
df = spark.sql('''
    CREATE OR REPLACE TEMPORARY VIEW tv_customers
    as 
    select * from json.`/Volumes/gizmobox/landing/operational_data/customers`
    '''
)
```

#### Spark DataFrame Reader API

```python
df = spark.read.format('json').options(inferSchema='true').load('/Volumes/gizmobox/landing/operational_data/customers')
display(df)
```

####  Reading Tables using spark.table.function

```python
df = spark.table('gizmobox.bronze.v_customers')

display(df)
```
