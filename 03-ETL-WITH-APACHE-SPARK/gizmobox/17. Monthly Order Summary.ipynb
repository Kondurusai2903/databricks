### Monthly Order Summary 

For Each of the Customer, Product the following summary per month

- total orders
- total items bought
- total amount spent

```python
%sql
select * from gizmobox.silver.customers;
```

```python
%sql
select * from gizmobox.silver.orders
```

```python
%sql
select date_format(transaction_timestamp, 'yyyy-MM') as order_month,
       customer_id,
       count(DISTINCT order_id) as total_orders,
       sum(quantity) as total_items_bought,
       SUM(price * quantity) as total_revenue
  from gizmobox.silver.orders 
  group by order_month, customer_id
```

```python
%sql
CREATE OR REPLACE TABLE gizmobox.gold.order_summary_monthly
as 
select date_format(transaction_timestamp, 'yyyy-MM') as order_month,
       customer_id,
       count(DISTINCT order_id) as total_orders,
       sum(quantity) as total_items_bought,
       SUM(price * quantity) as total_revenue
from gizmobox.silver.orders 
group by order_month, customer_id
```

```python
%sql
select * from gizmobox.gold.order_summary_monthly;
```
