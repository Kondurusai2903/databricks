###  Transform Membership Data 
 - Extract customer_id from the file_path
 - Write Transformed Data to the Silver Layer

#### Extract customer_id from the file_path

```python
%sql
select * from gizmobox.bronze.v_memberships;
```

```python
%sql
-- select get(split(path,'/'), size(split(path,'/')) - 1) as customer_id from gizmobox.bronze.v_memberships

select regexp_extract(path,'.*/([0-9]+)\\.png$',1) as customer_id, content as membership_card from gizmobox.bronze.v_memberships;
```

#### Write Transformed Data to the Silver Layer

```python
%sql
CREATE OR REPLACE TABLE gizmobox.silver.memberships
AS
select regexp_extract(path,'.*/([0-9]+)\\.png$',1) as customer_id, content as membership_card from gizmobox.bronze.v_memberships;
```

```python
%sql
select * from gizmobox.silver.memberships;
```
