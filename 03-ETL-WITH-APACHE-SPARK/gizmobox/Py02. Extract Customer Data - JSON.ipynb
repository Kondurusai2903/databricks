## Extract Data from the Customers Json File

- Query single File
- Query multiple Files using the wildcard Character
- Query all the Files in a Folder

```python
%fs ls /Volumes/gizmobox/landing/operational_data/customers
```

```python
%sql
DESCRIBE VOLUME gizmobox.landing.operational_data;
```

### Query Single Json File

```python
df = spark.read.format('json').load('/Volumes/gizmobox/landing/operational_data/customers/customers_2024_10.json')

df.printSchema()

display(df);
```

### Multiple Json Files using wildcard character

```python
df = spark.read.format('json').load('/Volumes/gizmobox/landing/operational_data/customers/customers_202*_*.json')

display(df)
```

Multiple Files using the Folder 

```python
df = spark.read.format('json').load('/Volumes/gizmobox/landing/operational_data/customers')

display(df)
```

```python
display(spark.read.json('/Volumes/gizmobox/landing/operational_data/customers'))
```

### Select File Metadata

- input_file_name() deprecated from 13.3 LTS onwards 

- _metadata is recommended for the future use 

```python
df_md = df.select('_metadata.file_path',"*")

display(df_md)
```

### Register Files in Unity Catalog using Views

```python
df_md.write.format('delta').mode('overwrite').saveAsTable('gizmobox.bronze.py_customers')
```

```python
df_md.writeTo('gizmobox.bronze.py_customers').createOrReplace();
```

```python
%sql
select * from gizmobox.bronze.py_customers
```

```python
%sql

use catalog gizmobox;
use schema bronze;

select * from v_customers;
```

### Create Temporary View

```python
%sql

CREATE OR REPLACE TEMPORARY VIEW vt_customers
as 
select * from json.`/Volumes/gizmobox/landing/operational_data/customers/`
```

```python
%sql
select * from vt_customers;
```

### Create a Global Temporary View

```python
%sql
CREATE OR REPLACE GLOBAL TEMPORARY VIEW gvt_customers
as 
select * from json.`/Volumes/gizmobox/landing/operational_data/customers/`
```

```python
%sql
select * from global_temp.gvt_customers;
```
