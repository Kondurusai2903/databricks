### Select File Metadata

- input_file_name() deprecated from 13.3 LTS onwards 

- _metadata is recommended for the future use 

```python
%sql
select input_file_name(),_metadata.file_path as file_path, * from json.`/Volumes/gizmobox/landing/operational_data/customers/`
```

### Register Files in Unity Catalog using Views

```python
%sql
CREATE OR REPLACE VIEW gizmobox.bronze.v_customers as 
select *,_metadata.file_path as file_path from json.`/Volumes/gizmobox/landing/operational_data/customers/`
```

```python
%sql
select * from gizmobox.bronze.v_customers
```

```python
%sql

use catalog gizmobox;
use schema bronze;

select * from v_customers;
```

### Create Temporary View

```python
%sql

CREATE OR REPLACE TEMPORARY VIEW vt_customers
as 
select * from json.`/Volumes/gizmobox/landing/operational_data/customers/`
```

```python
%sql
select * from vt_customers;
```

### Create a Global Temporary View

```python
%sql
CREATE OR REPLACE GLOBAL TEMPORARY VIEW gvt_customers
as 
select * from json.`/Volumes/gizmobox/landing/operational_data/customers/`
```

```python
%sql
select * from global_temp.gvt_customers;
```
