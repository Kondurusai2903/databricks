#### 2. Create External Table for Payments

```python
payment_schema = 'payment_id INTEGER,order_id INTEGER,payment_timestamp TIMESTAMP,payment_status INTEGER,payment_method STRING'
```

```python
from pyspark.sql.types import StructType,StructField,StringType,IntegerType,TimestampType

payment_schema = StructType([
    StructField('payment_id',IntegerType()),
    StructField('order_id',IntegerType()),
    StructField('payment_timestamp',TimestampType()),
    StructField('payment_status',IntegerType()),
    StructField('payment_method',StringType())
])
```

```python
df = spark.read.format('csv')\
        .option('header','true')\
        .schema(payment_schema)\
        .load('abfss://gizmobox@dbstorageacc123.dfs.core.windows.net/landing/external-data/payments')

display(df)
```

```python
df.writeTo('gizmobox.bronze.py_payments').createOrReplace();
```

```python
%sql
CREATE TABLE IF NOT EXISTS gizmobox.bronze.payments
(
    payment_id INTEGER,
    order_id INTEGER,
    payment_timestamp TIMESTAMP,
    payment_status INTEGER,
    payment_method STRING
)
USING csv
OPTIONS (
    header='true',
    delimiter=','
)
location 'abfss://gizmobox@dbstorageacc123.dfs.core.windows.net/landing/external-data/payments'
```
