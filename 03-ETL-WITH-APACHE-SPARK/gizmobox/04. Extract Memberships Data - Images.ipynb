
## Extract Data from the Membership - Image Files

- Query Membership Files using binaryFile Format
- Create a membership view in Bronze Schema

### 1. Query Memberships Folder using binaryFile Format


```python
%fs ls '/Volumes/gizmobox/landing/operational_data/memberships/'
```

```python
%sql
select * from binaryFile.`/Volumes/gizmobox/landing/operational_data/memberships/*/*.png`
```

### 2. Create a Membership View in Bronze Schema 

```python
%sql
CREATE OR REPLACE VIEW gizmobox.bronze.v_memberships
as 
select * from binaryFile.`/Volumes/gizmobox/landing/operational_data/memberships/*/*.png`
```

```python
%sql
select * from gizmobox.bronze.v_memberships;
```
