### User Defined Functions

- User Defined Functions(UDF) in Spark are the custom functions created by users extend the capabilities of spark SQL and pyspark.
- UDFs allow us to perform calculations or transformations to apply bussiness logic that are not possible with built in functions.
- You define the Function once and use it across mutliple queries
- SQL UDFs are recommended over Python UDFs due to better optimization

##### Syntax

**CREATE FUNCTION** catalog_name.schema_name.udf_name(param_name_data_type)

**RETURNS** return_type;

**RETURN** expression;

```python
%sql
CREATE OR REPLACE FUNCTION gizmobox.default.get_fullname(first_name string, last_name string)
RETURNS STRING
RETURN CONCAT(initcap(first_name),' ',initcap(last_name))
```

```python
%sql
select gizmobox.default.get_fullname('john','doe') as full_name
```

```python
%sql
DESC FUNCTION EXTENDED gizmobox.default.get_fullname
```

```python
%sql
CREATE OR REPLACE FUNCTION gizmobox.default.get_payment_status(payment_status int)
RETURNS STRING
RETURN CASE 
       WHEN payment_status = 1 THEN 'Success'
       WHEN payment_status = 2 THEN 'Pending'
       WHEN payment_status = 3 THEN 'Cancelled'
       WHEN payment_status = 4 THEN 'Failed'
END
```

```python
%sql
CREATE OR REPLACE FUNCTION gizmobox.default.upper_case(status string)
RETURNS STRING
RETURN upper(status)
```
