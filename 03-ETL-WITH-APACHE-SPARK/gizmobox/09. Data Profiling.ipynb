
### Data Profiling in Databricks

- Profile Data using UI
- Profile Data using DBUTILS package(dbutils.data.sumarize method)
- Profile Data Manually 
     - COUNT
     - COUNT_IF
     - MIN
     - MAX
     - WHERE CLAUSE

  

#### Profile Data using UI

```python
%sql
select * from gizmobox.bronze.v_customers;
```

#### Profile Data using DBUTILS package(dbutils.data.sumarize method)

```python
df = spark.table('gizmobox.bronze.v_customers')

dbutils.data.summarize(df)
```

#### Profile Data Manually
- Profile Data Manually 
     - COUNT
     - COUNT_IF
     - MIN
     - MAX
     - WHERE CLAUSE

```python
%sql
select count(customer_id) as unique_ids, count(email) as emails from gizmobox.bronze.v_customers
```

```python
%sql
select count_if(customer_id is NULL) as null_ids, count_if(email is NULL) as null_emails from gizmobox.bronze.v_customers
```

```python
%sql
select * from gizmobox.bronze.v_customers
where customer_id is NULL
```

```python
%sql
select count(*) as total_customers, count(DISTINCT customer_id) as total_unique_customers from gizmobox.bronze.v_customers
where customer_id is not null
```
