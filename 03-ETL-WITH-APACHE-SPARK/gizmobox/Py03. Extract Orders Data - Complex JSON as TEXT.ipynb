### Extract data from the Order JSON File

- Query order Files using the JSON Format
- Query order Files using the TEXT format
- Create a Orders View in the Bronze schema

#### 1. Query Order Files using the JSON Format

```python
df = spark.read.json('/Volumes/gizmobox/landing/operational_data/orders/orders_2024_10.json')

display(df)
```

```python
%sql 
select * from json.`/Volumes/gizmobox/landing/operational_data/orders/orders_2024_10.json`
```

```python
df = spark.read.format('text').load('/Volumes/gizmobox/landing/operational_data/orders/orders_2024_10.json')

display(df)
```

```python
%sql 
select * from text.`/Volumes/gizmobox/landing/operational_data/orders/orders_2024_10.json`
```

```python
df = spark.read.format('text').load('/Volumes/gizmobox/landing/operational_data/orders')

display(df)
```

```python
df = spark.read.text('/Volumes/gizmobox/landing/operational_data/orders')

display(df)
```

```python
df.writeTo('gizmobox.bronze.py_orders').createOrReplace()
```

```python
%sql 

create or replace view gizmobox.bronze.v_orders
as 
select * from text.`/Volumes/gizmobox/landing/operational_data/orders`
```

```python
%sql
select * from gizmobox.bronze.py_orders;
```
